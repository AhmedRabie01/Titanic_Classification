### Preprocessing the  data 

```python
train = Tr_data.drop(['Ticket','Cabin'],axis=1)
test = Tr_data.drop(['Ticket','Cabin'],axis=1)
```

```python
combine =[train,test]
```

```python
 # Analyze if Name feature can be engineered to extract titles and test correlation between titles and survival

for dataset in combine :
    dataset['Title'] = dataset.Name.str.extract('([A-Za-z]+)\.',expand=False)

pd.crosstab(train['Title'],test['Sex'])
```

```python
 # correlation between Rare & survived

for dataset in combine : 
    dataset ['Title'] = dataset ['Title'].replace(['Lady','Capt','Col','Countess','Don','Dr','lord','Major','Rev','Sir','Ms','Mrs','lord','Major'],'Rare')
 
    dataset['Title'] = dataset['Title'].replace('Mlle','Miss')
    dataset['Title'] = dataset['Title'].replace('Ms','Miss')
    dataset['Title'] = dataset['Title'].replace('Mne','Mrs')
    
    
    
train[['Title','Survived']].groupby(['Title'],as_index=False).mean()
    
```

```python
title_mapping={"Mr":1,"Miss":2,"Mrs":3,"Master":4,"Rare":5}
for dataset in combine:
    dataset['Title'] = dataset['Title'].map(title_mapping)
    dataset['Title'] = dataset['Title'].fillna(0)
    
train.head()
```

```python
train = train.drop(['Name','PassengerId'],axis=1)
test = test.drop(['Name'],axis=1)
combine = [train , test ]
train.shape,test.shape
```

```python
test.head()
```

```python
from sklearn.preprocessing import  LabelEncoder
labelEncoder = LabelEncoder()
train.iloc[ : , 2] = labelEncoder.fit_transform(train.iloc[ : , 2])
```

```python
train.head()
```

```python
 # correlation between Sex and Pclass
    
grid = sns.FacetGrid(train, row='Pclass', col='Sex', size=2.2, aspect=1.6)
grid.map(plt.hist, 'Age', alpha=.5, bins=20)
grid.add_legend()
```

```python
 # preparing an empty array to contain guessed Age values based on Pclass x Gender combinations.

guess_ages = np.zeros((2,3))
```

```python
for dataset in combine :
    for i in range(0,2) :
        for j in range(0,3) :
                guess_df = dataset[(dataset['Sex'] == i) & (dataset['Pclass'] == j+1)]['Age'].dropna()
                age_guess = guess_df.median()
                guess_ages[i,j] = float( age_guess/0.5 + 0.5 ) * 0.5
            
            
    for i in range(0, 2):
        for j in range(0, 3):
                dataset.loc[ (dataset.Age.isnull()) & (dataset.Sex == i) & (dataset.Pclass == j+1),'Age'] = guess_ages[i,j]
                dataset['Age'] = dataset['Age'].astype(float)

            
train.head()
```

```python
train['AgeBand'] = pd.cut(train['Age'], 5)
train[['AgeBand', 'Survived']].groupby(['AgeBand'], as_index=False).mean().sort_values(by='AgeBand', ascending=True)
```

```python
for dataset in combine:    
    dataset.loc[ dataset['Age'] <= 16, 'Age'] = 0
    dataset.loc[(dataset['Age'] > 16) & (dataset['Age'] <= 32), 'Age'] = 1
    dataset.loc[(dataset['Age'] > 32) & (dataset['Age'] <= 48), 'Age'] = 2
    dataset.loc[(dataset['Age'] > 48) & (dataset['Age'] <= 64), 'Age'] = 3
    dataset.loc[ dataset['Age'] > 64, 'Age']
train.head()
```

```python
train = train.drop(['AgeBand'], axis=1)
combine = [train, test]

```

```python
test['Age'] = train['Age']
```

```python
train.head()
```

```python
for dataset in combine:
    dataset['FamilySize'] = dataset['SibSp'] + dataset['Parch'] + 1

train[['FamilySize', 'Survived']].groupby(['FamilySize'], as_index=False).mean().sort_values(by='Survived', ascending=False)
```

```python
freq_port = train.Embarked.dropna().mode()[0]
freq_port
```

```python
for dataset in combine:
    dataset['Embarked'] = dataset['Embarked'].fillna(freq_port)
    
train[['Embarked', 'Survived']].groupby(['Embarked'], as_index=False).mean().sort_values(by='Survived', ascending=False)
```

```python
for dataset in combine:
    dataset['Embarked'] = dataset['Embarked'].map( {'S': 0, 'C': 1, 'Q': 2} ).astype(int)

train.head()
```

```python
test['Fare'].fillna(test['Fare'].dropna().median(), inplace=True)
test.head()
```

```python
train['FareBand'] = pd.qcut(train['Fare'], 4)
train[['FareBand', 'Survived']].groupby(['FareBand'], as_index=False).mean().sort_values(by='FareBand', ascending=True)
```

```python
for dataset in combine:
    dataset.loc[ dataset['Fare'] <= 7.91, 'Fare'] = 0
    dataset.loc[(dataset['Fare'] > 7.91) & (dataset['Fare'] <= 14.454), 'Fare'] = 1
    dataset.loc[(dataset['Fare'] > 14.454) & (dataset['Fare'] <= 31), 'Fare']   = 2
    dataset.loc[ dataset['Fare'] > 31, 'Fare'] = 3
    dataset['Fare'] = dataset['Fare'].astype(int)

train = train.drop(['FareBand'], axis=1)
combine = [train, test]
    
train.head(10)
```

```python
from sklearn.preprocessing import  LabelEncoder
labelEncoder = LabelEncoder()
test.iloc[ : , 3] = labelEncoder.fit_transform(test.iloc[ : , 3])
```

```python
X = train.iloc[:, 1:].values
```

```python
y = train.iloc[:, 0:1].values
```

### Train the Model &  predict

```python
 # Splitting the the trian set to Training set and Test set
    
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 41)
```

```python
 # KNN classifier 
    
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train,  y_train)
Y_pred = knn.predict(X_test)
acc_knn = round(knn.score(X_train, y_train) * 100, 2)
acc_knn
```

```python
 # Random forest classifier 

random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(X_train, y_train)
Y_pred = random_forest.predict(X_test)
random_forest.score(X_train,y_train)
acc_random_forest = round(random_forest.score(X_train, y_train) * 100, 2)
acc_random_forest
```
